```python
import pandas as pd
import pandas_datareader as pdr
```

```python
start = '1950-01-01'
Names = ['Fed BS', 'ECB BS', 'EURUSD', 'SPX']
df = pdr.DataReader(['WALCL', 'ECBASSETSW', 'DEXUSEU', 'SP500'], 'fred', start )
df.columns = Names
```

```python
df
```

```python
df = df.resample('W').last()
df = df.fillna(method = 'ffill')
df = df.dropna()
```

```python
df
```

```python
df['ECB in USD'] = df['ECB BS']*df['EURUSD']
```

```python
df['Tot BS'] = df['ECB in USD'] + df['Fed BS']
```

```python
df
```

```python
df['BS %Chg'] = df['Tot BS'].pct_change(12)
```

```python
df['SPX % chg'] = df['SPX'].pct_change(12)
```

```python
df['BS %Chg'].plot()
df['SPX % chg'].plot()
```
